```python
# Paths
feature_folder = '/home/gauss/Desktop/Revised_Data_Features/Pretrained baseline EfficientNetB0 Bbox Size 90/MHB/'
annot_folder = '/home/gauss/Desktop/Revised_Data_Features/EfficientNetB0 trained MHB Bbox Size 90/Plate annotations for MHB_NN/'
#annot_folder = '/4tbint/Plate_annotations_newMoAs/'

output_folder = '/home/gauss/Desktop/Revised_Data_Features/Pretrained baseline EfficientNetB0 Bbox Size 90/MHB_unormalized/'

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Get sorted feature and annotation files
feature_files = sorted([f for f in os.listdir(feature_folder) if f.endswith('.xlsx')])
annotation_files = sorted([f for f in os.listdir(annot_folder) if f.endswith('.xlsx')])

# Process each pair of feature and annotation files
for feature_file, annot_file in zip(feature_files, annotation_files):
    print(feature_file, annot_file)
    # Load data
    features = pd.read_excel(os.path.join(feature_folder, feature_file))
    annotation = pd.read_excel(os.path.join(annot_folder, annot_file))

    # Merge features and annotations
    merged_data = pd.concat([features, annotation], axis=1)
    
    
    # Save normalized data
    plate_name = os.path.splitext(feature_file)[0]  # Extract plate name without extension
    output_path = os.path.join(output_folder, f"{plate_name}.xlsx")
    merged_data.to_excel(output_path, index=False)

    print(f"Unnormalized data for {plate_name} saved to {output_path}.")

```

# Merge all the plates of all the combinations in one single dataframe 

```python
appended_data = []

unnorm_features = '/home/gauss/Desktop/Revised_Data_Features/Pretrained baseline EfficientNetB0 Bbox Size 90/MHB_unormalized/'

for feature_file in sorted(os.listdir(unnorm_features)):
    print(feature_file)
    unnorm_feature_data = pd.read_excel(os.path.join(unnorm_features, feature_file))
    
    #prefix_comb = feature_file.split("_")[1]
    prefix_comb = feature_file.split(".")[0]
    print(prefix_comb)

    if "A81" in prefix_comb:
        combo = "O3B"
        
    elif "A82" in prefix_comb:
        combo = 'KB'
    
    elif "Set25" in prefix_comb:
        combo = 'KW'
        
    elif "Set89" in prefix_comb:
        combo = 'O8W'
        
    elif "863" in prefix_comb:
        combo = 'MHB'
        
    combination = combo
    #print(combination)

    unnorm_feature_data.insert(0,'Combination', combination)
    
    #prefix_plate = feature_file.split("_")[1]
    #prefix_plate = prefix_comb.split("_")[1]
    prefix_plate = list(prefix_comb)
    #print(prefix_plate)

    unnorm_feature_data.insert(0,'Plate number', prefix_plate[1] )

    appended_data.append(unnorm_feature_data)
    
appended_data = pd.concat(appended_data)
# write DataFrame to an excel sheet 
appended_data.to_excel('/home/gauss/Desktop/Revised_Data_Features/Pretrained baseline EfficientNetB0 Bbox Size 90/MHB_unormalized/all_data.xlsx')
```

```python
import os
import pandas as pd

# Define the base path
base_path = "/home/gauss/Desktop/Revised_Data_Features/Pretrained baseline EfficientNetB0 Bbox Size 90/"
directories = ["KB_unormalized", "KW_unormalized", "MHB_unormalized", "O3B_unormalized", "O8W_unormalized"]

# Initialize an empty list to store dataframes
dataframes = []

# Loop through each directory
for dir_name in directories:
    dir_path = os.path.join(base_path, dir_name)
    all_data_file = os.path.join(dir_path, "all_data.xlsx")
    
    if os.path.exists(all_data_file):
        # Read the all_data file into a dataframe
        df = pd.read_excel(all_data_file)
        dataframes.append(df)
    else:
        print(f"File not found: {all_data_file}")

# Merge all dataframes
if dataframes:
    merged_df = pd.concat(dataframes, ignore_index=True)
    # Save the merged dataframe to a file
    output_file = os.path.join(base_path, "unnorm_efficientnetb0_all_filteredbbox.xlsx")
    merged_df.to_excel(output_file, index=False)
    print(f"Merged data saved to: {output_file}")
else:
    print("No dataframes to merge.")
```
